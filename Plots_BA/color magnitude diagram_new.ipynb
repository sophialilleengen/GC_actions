```python
#color magnitude diagram for the binary stellar system
fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111)

cax=ax.scatter(mbmv[1:],mv[1:],c=mass[1:],cmap='Spectral',edgecolor="None")
cbar = plt.colorbar(cax)
cbar.ax.get_yaxis().labelpad = 20
cbar.set_clim(0,2.4)

cbar.set_label('mass of star [M$_\odot$]', rotation=270,fontsize=14)
ax.set_xlabel('color B-V [mag]',fontsize=14)
ax.set_ylabel('absolute magnitude M$_\mathrm{v}$ [mag]',fontsize=14)
ax.set_ylim(30,-5)
#ax.set_title('color magnitude diagram',fontsize=2)
ax.spines['top'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)
ax.annotate('main sequence',xy=(1.3,10),xytext=(2.,10),arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('sub giant branch',xy=(0.6,3.8),xytext=(0.8,5),arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('red giant branch',xy=(1.0,0.0),xytext=(1.5,2.0),arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('quiscent\nHe burning',xy=(0.7,-1.),xytext=(-0.4,0),arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('early asymptotic\ngiant branch',xy=(1.3,-1.9),xytext=(0.0,-2.5),arrowprops=dict(facecolor='black', shrink=0.05))
#    ax.annotate('thermally pulsing \n asymptotic giant \n branch',xy=(1.6,-2.5),xytext=(1.9,-2.0),arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('blue stragglers',xy=(0.1,4),xytext=(-0.4,7),arrowprops=dict(facecolor='black', shrink=0.05)) #left to turning points
ax.annotate('white dwarfs',xy=(0.5,18),xytext=(0.5,23),arrowprops=dict(facecolor='black', shrink=0.05))

ax.tick_params(length=8,width=1.5,labelsize=12)
ax.tick_params(which='minor',length=4,width=1.5)
fig.tight_layout()
fig.savefig('/home/sophia/Bachelorarbeit-Entwurf/Plots/color_magnitude_diagram.png',format='png',dpi=400,bbox_inches='tight')
```
