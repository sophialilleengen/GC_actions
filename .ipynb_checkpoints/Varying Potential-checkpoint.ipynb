```python
M_BH = np.array([0, 100, 1000, 10000, 50000])
n = 1000
R = np.logspace(-3, 2, n)
pot1 = np.zeros(n)
pot2 = np.zeros(n)
pot1_wo = np.zeros(n)
pot2_wo = np.zeros(n)
for j in M_BH:
    MGE1 = MGE_potential(inputfilename = 'MGE_results_IMBH1', M_BH = j) 
    MGE2 = MGE_potential(inputfilename = 'MGE_results_IMBH2', M_BH = j)
    MGE3 = MGE_potential(inputfilename = 'MGE_results_noIMBH1', M_BH = j)
    MGE4 = MGE_potential(inputfilename = 'MGE_results_noIMBH2', M_BH = j)
    for i in range(n):
        pot1[i] = MGE1.potential(R[i])
        pot2[i] = MGE2.potential(R[i])
        pot1_wo[i] = MGE3.potential(R[i])
        pot2_wo[i] = MGE4.potential(R[i])
    plt.semilogx(R, pot1, '.', label ='1st SIM')
    plt.semilogx(R, pot2, '.', label ='2nd SIM')
    plt.semilogx(R, pot1_wo, '.', label ='3rd SIM')
    plt.semilogx(R, pot2_wo, '.', label ='4th SIM')
    plt.title('IMBH mass = ' + str(j) + '$M_\odot$')
    plt.legend(loc = 0, numpoints = 1)
    plt.show()
```
